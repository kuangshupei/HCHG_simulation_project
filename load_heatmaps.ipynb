```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
UVC_signal_energy_data = pd.read_csv("signal_energy_fiber_radius_100microns_Xenon_varyingFWHM.csv")
UVC_signal_peak_power_data = pd.read_csv("signal_peak_power_fiber_radius_100microns_Xenon_varyingFWHM.csv")
```

```python
frep_mks = 1e3  # repetition rate of input pulse [Hz]

UVC_signal_average_power_data = pd.DataFrame(columns=['length', 'pressure', 'average power'])
UVC_signal_average_power_data['length'] = UVC_signal_energy_data['length']
UVC_signal_average_power_data['pressure'] = UVC_signal_energy_data['pressure']
UVC_signal_average_power_data['average power'] = UVC_signal_energy_data['energy'] * frep_mks
```

```python
heatmap_UVC_signal_energy = UVC_signal_energy_data.pivot(index='pressure', columns='length', values='energy')

plt.figure(figsize=(10, 6))
plt.imshow(heatmap_UVC_signal_energy, aspect='auto', cmap='YlOrRd', origin='lower')

plt.colorbar(label='Signal energy [J]')
plt.title('Heatmap of integrated signal energy')
plt.xlabel('Length along fiber [mm]')
plt.ylabel('Pressure [Torr]')
plt.yticks(ticks=np.arange(0, len(heatmap_UVC_signal_energy.index)+100, 100))
plt.show()

heatmap_UVC_signal_average_power = UVC_signal_average_power_data.pivot(index='pressure', columns='length', values='average power')

plt.figure(figsize=(10, 6))
plt.imshow(heatmap_UVC_signal_average_power, aspect='auto', cmap='YlOrRd', origin='lower')

plt.colorbar(label='Average Power [W]')
plt.title('Heatmap of integrated signal average power')
plt.xlabel('Length along fiber [mm]')
plt.ylabel('Pressure [Torr]')
plt.yticks(ticks=np.arange(0, len(heatmap_UVC_signal_average_power.index)+100, 100))
plt.show()

heatmap_UVC_signal_peak_power = UVC_signal_peak_power_data.pivot(index='pressure', columns='length', values='peak power')

plt.figure(figsize=(10, 6))
plt.imshow(heatmap_UVC_signal_peak_power, aspect='auto', cmap='YlOrRd', origin='lower')

plt.colorbar(label='Peak Power [W]')
plt.title('Heatmap of integrated signal peak power')
plt.xlabel('Length along fiber [mm]')
plt.ylabel('Pressure [Torr]')
plt.yticks(ticks=np.arange(0, len(heatmap_UVC_signal_peak_power.index)+100, 100))
plt.show()
```
