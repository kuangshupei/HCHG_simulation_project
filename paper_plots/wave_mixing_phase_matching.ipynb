```python
import numpy as np
import matplotlib.pyplot as plt

def wave_mixing_efficiency(length_grid, phase_mismatch):
    eta = length_grid**2 * np.sinc((phase_mismatch * length_grid) / (2 * np.pi))**2
    return eta

phase_mismatches = [0, np.pi, 2*np.pi]
length_grid = np.linspace(0, 4, 50)

plt.figure(figsize=(8, 6))

for phase_mismatch in phase_mismatches:
    generated_intensity = wave_mixing_efficiency(length_grid, phase_mismatch)
    plt.plot(length_grid, generated_intensity, label=f'Δk = {phase_mismatch:.2f} rad/m')

plt.ylim(0, 2)
plt.xlim(0, 4)
plt.xlabel('Nonlinear Medium Length (L) [a.u.]', fontsize=14)
plt.ylabel('Generated Intensity [a.u.]', fontsize=14)
plt.title('Wave Mixing Efficiency vs. Nonlinear Medium Length', fontsize=16)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.legend(title='Phase Mismatch (Δk)', fontsize=12, title_fontsize=13, loc='upper right')

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# define the function to calculate wave mixing efficiency
def wave_mixing_efficiency(length_grid, phase_mismatch):
    eta = length_grid**2 * np.sinc((phase_mismatch * length_grid) / (2 * np.pi))**2
    return eta

# phase mismatches
phase_mismatches = [0, np.pi, 2*np.pi]
length_grid = np.linspace(0, 4, 500)

sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)

colors = sns.color_palette("Blues", len(phase_mismatches))

plt.figure(figsize=(8, 6))

labels = {
    0: r'$\Delta k = 0$',
    np.pi: r'$\Delta k = \pi$',
    2*np.pi: r'$\Delta k = 2\pi$'
}

for phase_mismatch, color in zip(phase_mismatches, colors):
    generated_intensity = wave_mixing_efficiency(length_grid, phase_mismatch)
    plt.plot(length_grid, generated_intensity, label=labels[phase_mismatch], color=color, linewidth=2)

plt.ylim(0, 2)
plt.xlim(0, 4)
plt.xlabel('Length of Nonlinear Medium (a.u.)', fontsize=14)
plt.ylabel('Generated Intensity (a.u.)', fontsize=14)

plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.legend(title='Phase Mismatch (Δk)', fontsize=12, title_fontsize=14, loc='upper right')

plt.tight_layout()
plt.savefig("fig3.png") 
plt.show()
```
