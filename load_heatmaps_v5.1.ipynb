```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
TH_energy_dataframe = pd.read_csv("third_harmonic_energy_dataframe_v5.1[Xenon,radius=100microns,pressure=upto1bar,length=upto500mm,useautocorrelation].csv")
TH_peak_power_dataframe = pd.read_csv("third_harmonic_peak_power_dataframe_v5.1[Xenon,radius=100microns,pressure=upto1bar,length=upto500mm,useautocorrelation].csv")
FH_energy_dataframe = pd.read_csv("fourth_harmonic_energy_dataframe_v5.1[Xenon,radius=100microns,pressure=upto1bar,length=upto500mm,useautocorrelation].csv")
FH_peak_power_dataframe = pd.read_csv("fourth_harmonic_peak_power_dataframe_v5.1[Xenon,radius=100microns,pressure=upto1bar,length=upto500mm,useautocorrelation].csv")
```

```python
heatmap_TH_energy = TH_energy_dataframe.pivot(index='gas pressure', columns='fiber length', values='third harmonic energy')

plt.figure(figsize=(10, 6))
plt.imshow(heatmap_TH_energy, aspect='auto', cmap='YlOrRd', origin='lower')

plt.colorbar(label='third harmonic energy [J]')
plt.xlabel('fiber length [mm]')
plt.ylabel('gas pressure [bar]')
plt.show()

heatmap_TH_peak_power = TH_peak_power_dataframe.pivot(index='gas pressure', columns='fiber length', values='third harmonic peak power')

plt.figure(figsize=(10, 6))
plt.imshow(heatmap_TH_peak_power, aspect='auto', cmap='YlOrRd', origin='lower')

plt.colorbar(label='third harmonic peak power [W]')
plt.xlabel('fiber length [mm]')
plt.ylabel('gas pressure [bar]')
plt.show()
```

```python
heatmap_FH_energy = FH_energy_dataframe.pivot(index='gas pressure', columns='fiber length', values='fourth harmonic energy')

plt.figure(figsize=(10, 6))
plt.imshow(heatmap_FH_energy, aspect='auto', cmap='YlOrRd', origin='lower')

plt.colorbar(label='fourth harmonic energy [J]')
plt.xlabel('fiber length [mm]')
plt.ylabel('gas pressure [bar]')
plt.show()

heatmap_FH_peak_power = FH_peak_power_dataframe.pivot(index='gas pressure', columns='fiber length', values='fourth harmonic peak power')

plt.figure(figsize=(10, 6))
plt.imshow(heatmap_FH_peak_power, aspect='auto', cmap='YlOrRd', origin='lower')

plt.colorbar(label='fourth harmonic peak power [W]')
plt.xlabel('fiber length [mm]')
plt.ylabel('gas pressure [bar]')
plt.show()
```

For fiber radius 100.0 microns, a total of 0.1131 mJ is needed, of which the fundamental pulse has 0.09425 mJ, the second harmonic pulse has 0.01885 mJ.

```python
total_EPP = 0.1131 * 1e-3  # [J]
FWHM = 0.03  # [ps]
```

```python
maximum_energy_index = TH_energy_dataframe['third harmonic energy'].idxmax()
third_harmonic_maximum_energy = TH_energy_dataframe.loc[maximum_energy_index, 'third harmonic energy']
corresponding_peak_power = TH_peak_power_dataframe.loc[maximum_energy_index, 'third harmonic peak power']
corresponding_gas_pressure = TH_energy_dataframe.loc[maximum_energy_index, 'gas pressure']
corresponding_fiber_length = TH_energy_dataframe.loc[maximum_energy_index, 'fiber length']

print(f"The maximum energy conversion efficiency (for third harmonic) is {(third_harmonic_maximum_energy/total_EPP)*100}% at pressure {corresponding_gas_pressure}bar and length {corresponding_fiber_length}mm.")
print(f"The peak power conversion efficiency (for third harmonic) is {(corresponding_peak_power/(total_EPP/(FWHM*1e-12)))*100}% at pressure {corresponding_gas_pressure}bar and length {corresponding_fiber_length}mm.")
```

```python
maximum_energy_index = FH_energy_dataframe['fourth harmonic energy'].idxmax()
fourth_harmonic_maximum_energy = FH_energy_dataframe.loc[maximum_energy_index, 'fourth harmonic energy']
corresponding_peak_power = FH_peak_power_dataframe.loc[maximum_energy_index, 'fourth harmonic peak power']
corresponding_gas_pressure = FH_energy_dataframe.loc[maximum_energy_index, 'gas pressure']
corresponding_fiber_length = FH_energy_dataframe.loc[maximum_energy_index, 'fiber length']

print(f"The maximum energy conversion efficiency (for fourth harmonic) is {(fourth_harmonic_maximum_energy/total_EPP)*100}% at pressure {corresponding_gas_pressure}bar and length {corresponding_fiber_length}mm.")
print(f"The peak power conversion efficiency (for fourth harmonic) is {(corresponding_peak_power/(total_EPP/(FWHM*1e-12)))*100}% at pressure {corresponding_gas_pressure}bar and length {corresponding_fiber_length}mm.")
```
